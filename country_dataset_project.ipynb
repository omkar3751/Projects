# DATA CLEANING

```python
df.drop(['id','alpha-2','iso_3166-2','intermediate-region','intermediate-region-code','Unnamed: 13','population.1'],
        axis=1,inplace=True) # Drop the column
```

```python
df.rename(columns={'name':'country_name'},inplace=True)                                           # Rename the column name
```

```python
df.loc[df.region.isna(),'region'] = 'Americas'                                                  # Null values to fill
```

```python
df.loc[df['sub-region'].isna(),'sub-region'] = 'Latin America and the Caribbean'
```

```python
df.loc[df['region-code'].isna(),'region-code'] = '19.0'
```

```python
df.loc[df['sub-region-code'].isna(),'sub-region-code'] = '419.0'
```

# DATA EXPLORATION / VISUALIZATION

```python
plt.bar(df['region'],df['population'],label='region')                                       # region by population bar graph
plt.xlabel('Regions',color='red',fontsize=12)
plt.ylabel('Population',color='red',fontsize=12)
plt.legend()
plt.show()
```

```python
plt.hist(df['region'],bins=9)                                                                 # region wise frequency
plt.xlabel('Regions',color='red',fontsize=16)
plt.ylabel('Frequency',color='red',fontsize=16)
```

```python
sns.countplot(df['sub-region'],palette='rainbow')                                         # count the Sub-regions
plt.xlabel('Count',color='purple',fontsize=14)
plt.ylabel('Sub-region',color='purple',fontsize=14)
```

```python
sns.barplot(df['region'],palette='rocket')                                                # barplot of region
plt.ylabel("Region",color = 'green',fontsize=15)
```

```python
sns.jointplot(x=df['region'],y=df['sub-region'])                                       # Region & Sub-region joint plot
```
